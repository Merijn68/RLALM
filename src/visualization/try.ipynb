```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Sample data: inflows and outflows for the next 30 years
years = np.arange(1, 31)
inflows = np.random.randint(100, 1000, size=30)
outflows = np.random.randint(50, 500, size=30)

# Calculate the difference between inflows and outflows
differences = inflows - outflows

# Combine data for plotting
data = {'Years': years,
        'Inflows': inflows,
        'Outflows': outflows,
        'Positive Difference': np.maximum(differences, 0),
        'Negative Difference': np.minimum(differences, 0)}

# Create a stylish stacked bar plot using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Years', y='Inflows', color='skyblue', label='Inflows')
sns.barplot(data=data, x='Years', y='Outflows', color='coral', label='Outflows')
sns.barplot(data=data, x='Years', y='Positive Difference', color='lime', label='Positive Difference')
sns.barplot(data=data, x='Years', y='Negative Difference', color='red', label='Negative Difference')

# Customize the plot
plt.title('Inflows, Outflows, and Differences Over Time')
plt.xlabel('Years')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

tenors = [1, 2, 3, 5, 10, 20]
yield_curve = [0.02, 0.025, 0.03, 0.035, 0.04, 0.045]

fig, ax = plt.subplots(figsize=(10, 6))

sns.lineplot(x=tenors, y=yield_curve, marker="o", ax=ax, label="Yield Curve")

distribution_data = {
    tenor: np.random.normal(yield_val, 0.005, 100) for tenor, yield_val in zip(tenors, yield_curve)
}

sns.stripplot(data=[distribution_data[tenor] for tenor in tenors], jitter=True, alpha=0.5, ax=ax)

ax.set_xlabel("Tenor")
ax.set_ylabel("Yield")
ax.set_title("Yield Curve and Distribution by Tenor")

ax.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Example data
tenors = [1, 2, 3, 5, 10, 20]
mortgages = [50, 75, 100, 125, 150, 175]
funding = [40, 70, 95, 120, 145, 170]

# Calculate differences at each tenor point
differences = [m - f for m, f in zip(mortgages, funding)]

# Create a DataFrame
data = pd.DataFrame({'Tenors': tenors, 'Mortgages': mortgages, 'Funding': funding, 'Differences': differences})

# Create a figure with subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Create grouped bar plots for mortgages and funding
bar_width = 0.35
ax1.bar(data['Tenors'], data['Mortgages'], bar_width, label='Mortgages', color='blue')
ax1.bar(data['Tenors'] + bar_width, data['Funding'], bar_width, label='Funding', color='orange')

# Set labels and title for the first plot
ax1.set_xlabel('Tenor')
ax1.set_ylabel('Count')
ax1.set_title('Counts of Mortgages and Funding per Tenor Point')
ax1.legend()

# Create a table with differences for each tenor point
table_data = [['Total Mortgages', sum(mortgages)],
              ['Total Funding', sum(funding)],
              ['Total Difference', sum(differences)]]

for tenor, diff in zip(tenors, differences):
    table_data.append([f'Difference at Tenor {tenor}', diff])

table = ax2.table(cellText=table_data, loc='center', cellLoc='left')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)

# Remove axes for the second plot
ax2.axis('off')

# Set the layout
plt.tight_layout()

# Show the plot
plt.show()


```
